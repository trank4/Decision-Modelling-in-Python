```python
#libraries
from ortools.linear_solver import pywraplp as glp
```

```python
#part 1:
#inputs
# each row (sublist) corresponds to a single student
# each element in a row (sublist) indicates the student's preference for a project

preference = [[2,5,3,2,4,4,5],
              [1,5,1,2,2,4,5],
              [2,5,4,4,5,5,3],
              [2,5,3,1,5,4,4],
              [2,4,3,1,5,5,5],
              [5,5,2,2,2,3,5],
              [2,5,2,3,4,5,5],
              [2,5,3,1,3,5,4],
              [1,5,2,3,4,5,1],
              [3,5,4,2,3,5,5],
              [2,5,2,2,2,2,5],
              [1,5,1,2,2,5,4],
              [1,4,1,2,1,4,5],
              [1,5,1,2,2,5,4],
              [1,5,1,1,4,5,4],
              [2,5,3,2,4,5,3],
              [1,5,1,2,4,3,4],
              [3,4,2,1,3,5,4],
              [5,4,3,5,5,4,5],
              [5,5,1,4,5,5,1],
              [2,5,1,1,4,5,4],
              [1,4,4,5,2,2,3],
              [4,5,1,1,5,5,5],
              [1,4,1,3,5,2,3],
              [1,3,1,2,2,4,5],
              [2,5,4,4,5,5,3],
              [2,4,3,2,3,5,4],
              [3,5,3,1,4,5,5]]
projpstu=1
stupproj=4
```

```python
#Create model
model = glp.Solver('student-project',glp.Solver.GLOP_LINEAR_PROGRAMMING)
```

```python
#Decision variables:
assign=list(range(len(preference)))
for s in range(len(preference)):
    assign[s]=list(range((len(preference[1]))))
    for p in range(len(preference[1])):
        assign[s][p]=model.NumVar(0,1,'student %d - project %d'%(s,p))
```

```python
#Objective Function
TotalScore=model.Objective()
TotalScore.SetMaximization()
for s in range(len(preference)):
    for p in range(len(preference[1])):
        TotalScore.SetCoefficient(assign[s][p],preference[s][p])
```

```python
#Constraints on "1 student is assigned 1 project"
projpstu_constr = list(range(len(preference)))
for s in range(len(preference)):
    projpstu_constr[s]= model.Constraint(projpstu,projpstu,'# of project for student %d'%s)
    for p in range(len(preference[1])):
        projpstu_constr[s].SetCoefficient(assign[s][p],1)
#Constraints on "1 project is assigned 4 students"
stupproj_constr = list(range(len(preference[1])))
for p in range(len(preference[1])):
    stupproj_constr[p]= model.Constraint(stupproj,stupproj,'# of students for project %d'%p)
    for s in range(len(preference)):
        stupproj_constr[p].SetCoefficient(assign[s][p],1)
```

```python
#Solve model and display result
statusls = ['Optimal','Feasible','Infeasible','Unbounded','Abnormal','Not Solved']
status = model.Solve()
print('Solution status:', statusls[status])
print('Number of variables:', model.NumVariables())
print('Number of constraints:',model.NumConstraints())
print('Total score: %.0f'%(TotalScore.Value()))
```

```python
#Display the Assignment
for s in range(len(preference)):
    print('student %d'%s)
    for p in range(len(preference[1])):
        print('project %d'%p,assign[s][p].solution_value())
```
